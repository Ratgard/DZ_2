```python
data = pd.read_csv('KAG_conversion_data.csv')
data.head(5)
```

```python
data['interest'] = data['interest'].astype('object')
```

```python
import matplotlib.pyplot as plt
```

```python
age_conversion = data[['age', 'Total_Conversion', 'Approved_Conversion']]\
    .groupby('age').sum()
#age_conversion
```

Конверсии по полу и возрасту (График)

```python
m_gender = data.loc[data['gender'].isin(['M'])]
#m_gender
m_age_conversion = m_gender[['age', 'gender', 'Approved_Conversion']] \
                            .groupby('age').sum()
#m_age_conversion

f_gender = data.loc[data['gender'].isin(['F'])]
#f_gender
f_age_conversion = f_gender[['age', 'Approved_Conversion']] \
                           .groupby('age').sum()
#f_age_conversion
```

```python
plt.plot(m_age_conversion.index, m_age_conversion['Approved_Conversion'])
plt.title('Конверсии по полу и возрасту')
plt.xlabel('возраст')
plt.ylabel('Конверсии')


plt.plot(f_age_conversion.index, f_age_conversion['Approved_Conversion'])
plt.title('Конверсии по полу и возрасту')
plt.xlabel('возраст')
plt.ylabel('Конверсии')
```

График показывающий Approved_Conversion по полу и возрасту. Среди женщий самый низкий выкуп у аудитории 35-39,Самый низкий у мужчик 40-44. Из гипотез мы можем установить корректировки по полу и возруста для показа рекламы в фейсбуке на те или иные группы, тем самым повлиять на маркетинговые показатели. Но прежде всего надо сравнить количество трафика, возможно оно сопоставимо с количеством конверсий. Либо высчитать качественный показатель, а именно %CR в заданном разрезе.

График сравнение CR% в разрезе пол/возраст

```python
m_gender_cr = m_gender[['age', 'Clicks', 'Total_Conversion']] \
                            .groupby('age').sum()
m_gender_cr['CR%']=m_gender_cr['Total_Conversion'] / m_gender_cr['Clicks']  * 100
m_gender_cr_round = m_gender_cr['CR%'].round(2).reset_index()
#m_gender_cr_round

f_gender_cr = f_gender[['age', 'Clicks', 'Total_Conversion']] \
                            .groupby('age').sum()
f_gender_cr['CR%']=f_gender_cr['Total_Conversion'] / f_gender_cr['Clicks']  * 100
f_gender_cr_round = f_gender_cr['CR%'].round(2).reset_index()
#f_gender_cr_round
```

```python
plt.plot(m_gender_cr_round['age'], m_gender_cr_round['CR%'])
plt.title('CR% по полу и возрасту')
plt.xlabel('возраст')
plt.ylabel('Конверсии')

plt.plot(f_gender_cr_round['age'], f_gender_cr_round['CR%'])
plt.title('CR% по полу и возрасту')
plt.xlabel('возраст')
plt.ylabel('Конверсии')
```

В целом мы видим падение конверсии по М и Ж с увеличением возраста. Информации о продукте нет. Можно сказать что отрабатывает лучше всего ЦА 30-44. Опираясь на предыдущий график можно предположить следующее: т.к. количество конверсий у аудитории 45-49 больше чем у 35-39 и 40-44 (опираясь на предыдущий график), а качественный показатель CR у ца 45-49 ниже, то можно сказать что трафика по ца 45-49 больше чем по отдельно взятым 35-39 и 40-44. Можно посмотреть срезы в разрезах кампаний (в общем) затем углубиться и посмотреть по интересам и таким образом скорректировать таргетинги и/или ставки. То есть таким образом мы урежем часть не целевой аудитории в 45-49, конверсия в следствии будет выше, при том что количество трафика уменьшится, а высвобожденные деньги лучше будет перераспределить на более эффективные.

Столбчатая диаграмма. Топ 10 интересов по количеству конверсий

```python
interest_conversion_top10= data[['interest', 'Approved_Conversion']].groupby('interest').sum().sort_values('Approved_Conversion', ascending=False).head(10)
interest_conversion_top10.plot(kind='bar', title= 'Топ 10 интересов по продажам', 
                            xlabel= 'Категория',
                            ylabel= 'Количество')
for k, v in enumerate(interest_conversion_top10.values):
    plt.text(k, v, int(v), horizontalalignment='center', verticalalignment='bottom', fontdict={'size':14})
```

К сожалению тут тоже нет названий даже. Можно сказать что первые 2 явно лидируют. Скорее всего первые 2-3 интереса являются более широкими - то есть большой охват и как следствие много конверсий, но для того что бы сказать об эффективности я бы посмотрел хотя бы в общем CR% в сравнении с другими и количество трафика. Либо эти интересы являются максимально целевыми, а остальные косвенно связанные с продуктом (околоцелевые) и трафик там более холодный.

Данный Boxplot показывает выбросы, 200%, 100%, 77%, есть длинный хвост %конверсии, это показатели явно влияющие на средние, которые могут привести к некорректным выводам в дальнейшем при работе с данными.

```python
data_cr_25 = data.loc[data['CR%']<=25]
#data_cr_25
```
